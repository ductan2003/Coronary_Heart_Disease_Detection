```python
import json
import matplotlib.pyplot as plt
import numpy as np
import matplotlib


matplotlib.rc('font', size=14)  # Set default font size for all elements
matplotlib.rc('axes', titlesize=18)  # Set title font size
matplotlib.rc('axes', labelsize=16)  # Set axes label font size
matplotlib.rc('xtick', labelsize=14)  # Set x-tick label font size
matplotlib.rc('ytick', labelsize=14)  # Set y-tick label font size
matplotlib.rc('legend', fontsize=14)  # Set legend font size

def get_data(path):
    with open(path) as f1:
        data1 = json.load(f1)
    return data1

def get_detail(data1):
    indexes1 = [entry['step'] for entry in data1]
    num_nodes1 = [entry['num_nodes'] for entry in data1]
    edges1 = [entry['num_edges'] for entry in data1]
    return indexes1, num_nodes1, edges1

PGCN1 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/PGCN/Thesis_log/Env_Change/24-12-09-02:17:30/log_until_8750.json")
PGCN2 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/PGCN/Thesis_log/Env_Change/24-12-09-02:17:30/log_until_17200.json")
PGCN3 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/PGCN/Thesis_log/Env_Change/24-12-09-02:17:30/log_until_25180.json")
PGCN4 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/PGCN/Thesis_log/Env_Change/24-12-09-02:17:30/log_until_37360.json")

GNG1 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/MyGNG/theta20/8750.json")
GNG2 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/MyGNG/theta20/17200.json")
GNG3 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/MyGNG/theta20/25180.json")
GNG4 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/MyGNG/theta20/37360.json")

LLCS1 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/LLCS/Thesis_log/Env_Change/24-12-09-02:40:24/log_until_8750.json")
LLCS2 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/LLCS/Thesis_log/Env_Change/24-12-09-02:40:24/log_until_17200.json")
LLCS3 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/LLCS/Thesis_log/Env_Change/24-12-09-02:40:24/log_until_25180.json")
LLCS4 = get_data("/Users/tannguyen/Coronary_Heart_Disease_Detection/LLCS/Thesis_log/Env_Change/24-12-09-02:40:24/log_until_37360.json")

indexes1, num_nodes1, edges1 = get_detail(PGCN1)
indexes2, num_nodes2, edges2 = get_detail(PGCN2)
indexes3, num_nodes3, edges3 = get_detail(PGCN3)
indexes4, num_nodes4, edges4 = get_detail(PGCN4)

indexes5, num_nodes5, edges5 = get_detail(GNG1)
indexes6, num_nodes6, edges6 = get_detail(GNG2)
indexes7, num_nodes7, edges7 = get_detail(GNG3)
indexes8, num_nodes8, edges8 = get_detail(GNG4)

indexes9, num_nodes9, edges9 = get_detail(LLCS1)
indexes10, num_nodes10, edges10 = get_detail(LLCS2)
indexes11, num_nodes11, edges11 = get_detail(LLCS3)
indexes12, num_nodes12, edges12 = get_detail(LLCS4)


# Set line thickness
line_width = 2.0

# Define colors for each dataset
colors = {
    "PGCN": "blue",
    "GNG": "green",
    "LLCS": "orange"
}

# Plotting: Num nodes
plt.figure(figsize=(15, 5))

# Plot all environments for PGCN
plt.plot(indexes1, num_nodes1, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")  # No legend for individual lines
plt.plot(indexes2, num_nodes2, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes3, num_nodes3, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes4, num_nodes4, color=colors["PGCN"], linewidth=line_width, label="PGCN-300")

# Plot all environments for GNG
plt.plot(indexes5, num_nodes5, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes6, num_nodes6, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes7, num_nodes7, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes8, num_nodes8, color=colors["GNG"], linewidth=line_width, label="GNG-20")

# Plot all environments for LLCS
plt.plot(indexes9, num_nodes9, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes10, num_nodes10, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes11, num_nodes11, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes12, num_nodes12, color=colors["LLCS"], linewidth=line_width, label="LLCS")

# Customize the plot
plt.xlabel("Bước")
plt.ylabel("Số lượng nút")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plotting: Edges
plt.figure(figsize=(15, 5))

# Plot all environments for PGCN
plt.plot(indexes1, edges1, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes2, edges2, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes3, edges3, color=colors["PGCN"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes4, edges4, color=colors["PGCN"], linewidth=line_width, label="PGCN-300")

# Plot all environments for GNG
plt.plot(indexes5, edges5, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes6, edges6, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes7, edges7, color=colors["GNG"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes8, edges8, color=colors["GNG"], linewidth=line_width, label="GNG-20")

# Plot all environments for LLCS
plt.plot(indexes9, edges9, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes10, edges10, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes11, edges11, color=colors["LLCS"], linewidth=line_width, label="_nolegend_")
plt.plot(indexes12, edges12, color=colors["LLCS"], linewidth=line_width, label="LLCS")

# Customize the plot
plt.xlabel("Bước")
plt.ylabel("Số lượng cạnh")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
