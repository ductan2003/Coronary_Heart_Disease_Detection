```python
import os

def update_labels_in_csv(folder_path):
    # Iterate through all files in the specified folder
    for filename in os.listdir(folder_path):
        # Check if the file is a CSV
        if filename.endswith('.csv'):
            # Split the filename to extract the type
            parts = filename.split('_')
            if len(parts) >= 3:
                timestamp, label, file_type = parts

                # Check if the type is "Resting-Normal"
                if file_type == "Resting-Normal.csv":
                    # Update the label to 1
                    new_label = '0'
                    
                    # Create the new filename with the updated label
                    new_filename = f"{timestamp}_{new_label}_{file_type}"

                    # Full paths for the old and new filenames
                    old_file_path = os.path.join(folder_path, filename)
                    new_file_path = os.path.join(folder_path, new_filename)

                    # # Rename the file
                    os.rename(old_file_path, new_file_path)
                    print(f"Renamed '{filename}' to '{new_filename}'")

# Usage
folder_path = './Dataset1/ECG/'  # Replace with your folder path
update_labels_in_csv(folder_path)
```
