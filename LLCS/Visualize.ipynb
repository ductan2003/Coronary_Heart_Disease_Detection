```python
import json
import matplotlib.pyplot as plt
```

```python
file_path = './Dataset3.json'
with open(file_path, 'r') as f:
    data = json.load(f)
```

```python
file_path = './dataset3_GNG.json'
with open(file_path, 'r') as f:
    dataGNG = json.load(f)
```

```python
indices = [entry['index'] for entry in data]
num_nodes = [entry['num_nodes'] for entry in data]
edges = [entry['edge'] for entry in data]
```

```python
indicesGNG = [entry['index'] for entry in dataGNG]
num_nodesGNG = [entry['num_nodes'] for entry in dataGNG]
edgesGNG = [entry['edge'] for entry in dataGNG]
```

```python
indices = indices[:250]
num_nodes = num_nodes[:250]
edges = edges[:250]
indicesGNG = indicesGNG[:250]
num_nodesGNG = num_nodesGNG[:250]
edgesGNG = edgesGNG[:250]
```

```python
import matplotlib.pyplot as plt

# Assuming `indices`, `num_nodes`, `edges` are from the first file
# and `indicesGNG`, `num_nodesGNG`, `edgesGNG` are from the second file

plt.figure(figsize=(12, 8))

# Plot num_nodes for both files
plt.subplot(2, 1, 1)
plt.plot(indices, num_nodes, label='LLCS - num_nodes', marker='o', color='blue', linewidth=0.1)
plt.plot(indicesGNG, num_nodesGNG, label='GNG - num_nodes', marker='o', color='orange', linewidth=0.1)
plt.xlabel('Index')
plt.ylabel('Number of Nodes')
plt.title('Change in Number of Nodes Over Time')
plt.legend()  # Show legend to distinguish lines
plt.grid(True)

# Plot edge for both files
plt.subplot(2, 1, 2)
plt.plot(indices, edges, label='File 1 - edge', marker='o', color='blue')
plt.plot(indicesGNG, edgesGNG, label='File 2 - edge', marker='o', color='orange')
plt.xlabel('Index')
plt.ylabel('Edge Value')
plt.title('Change in Edge Value Over Time')
plt.legend()  # Show legend to distinguish lines
plt.grid(True)

# Show the plot with a tight layout
plt.tight_layout()
plt.show()
```
