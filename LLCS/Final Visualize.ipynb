```python
import json
import matplotlib.pyplot as plt

# Load the JSON data
with open('./dataset3_GNG_2.json') as f1, open('./Dataset3_1730574579.772365.json') as f2:
    data1 = json.load(f1)
    data2 = json.load(f2)

# Prepare data for plotting
indexes1 = [entry['index'] for entry in data1]
num_nodes1 = [entry['num_nodes'] for entry in data1]
edges1 = [entry['edge'] for entry in data1]

indexes2 = [entry['index'] for entry in data2]
num_nodes2 = [entry['num_nodes'] for entry in data2]
edges2 = [entry['edge'] for entry in data2]

# Plotting
plt.figure(figsize=(14, 7))

# Plot for num_nodes
plt.subplot(2, 1, 1)
plt.plot(indexes1, num_nodes1, label="num_nodes (dataset3_GNG)", color='blue')
plt.plot(indexes2, num_nodes2, label="num_nodes (Dataset3)", color='orange')
# plt.xlim(0, 3000)
# plt.ylim(0, 40)
plt.xlabel("Step")
plt.ylabel("Number of Nodes")
plt.title("Change of num_nodes over Step")
plt.legend()

# Plot for edge
plt.subplot(2, 1, 2)
plt.plot(indexes1, edges1, label="edge (dataset3_GNG)", color='blue')
plt.plot(indexes2, edges2, label="edge (Dataset3)", color='orange')
# plt.xlim(0, 3000)
# plt.ylim(0, 70)
plt.xlabel("Step")
plt.ylabel("Edge Count")
plt.title("Change of edge over step")
plt.legend()

plt.tight_layout()
plt.show()
```
