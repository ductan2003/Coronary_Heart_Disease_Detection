```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Load the JSON data
with open('./Dataset3_1730658594.76685.json') as f1, open('./Dataset3_1730658931.958479.json') as f2:
    data1 = json.load(f1)
    data2 = json.load(f2)

# Prepare data for plotting
indexes1 = [entry['index'] for entry in data1]
num_nodes1 = [entry['num_nodes'] for entry in data1]
edges1 = [entry['edge'] for entry in data1]
mse_values1 = [entry["mse"] for entry in data1]


indexes2 = [entry['index'] for entry in data2]
num_nodes2 = [entry['num_nodes'] for entry in data2]
edges2 = [entry['edge'] for entry in data2]
mse_values2 = [entry["mse"] for entry in data2]


# Plotting
plt.figure(figsize=(14, 7))

# Plot for num_nodes
plt.subplot(2, 1, 1)
plt.plot(indexes1, num_nodes1, label="num_nodes (dataset3_withAct)", color='blue')
plt.plot(indexes2, num_nodes2, label="num_nodes (Dataset3_6label)", color='orange')
# plt.xlim(0, 3000)
plt.ylim(0, 40)
plt.xlabel("Step")
plt.ylabel("Number of Nodes")
plt.title("Change of num_nodes over Step")
plt.legend()
plt.grid(True)
plt.show()

plt.subplot(2, 1, 1)
plt.plot(indexes1, mse_values1, label="num_nodes (dataset3_withAct)", color='blue')
plt.plot(indexes2, mse_values2, label="num_nodes (Dataset3)", color='orange')
plt.ylim(0, 0.5)
plt.xlabel("Time Step (Index)")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("MSE Through Time")
plt.grid(True)
plt.legend()
plt.show()


# Plot for edge
plt.subplot(2, 1, 2)
plt.plot(indexes1, edges1, label="edge (dataset3_withAct)", color='blue')
plt.plot(indexes2, edges2, label="edge (Dataset3)", color='orange')
# plt.xlim(0, 3000)
# plt.ylim(0, 70)
plt.xlabel("Step")
plt.ylabel("Edge Count")
plt.title("Change of edge over step")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
