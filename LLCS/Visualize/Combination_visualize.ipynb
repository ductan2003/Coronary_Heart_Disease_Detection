```python
import os
import glob
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
def load_and_save_logs(folder_path, output_folder):
    """
    This function loads all JSON files in the specified folder with the naming convention 'log_until_<step>.json',
    combines their data, and saves each of the four graphs (nodes, edges, MSE, and success rate) as separate images.
    """
    # Initialize an empty list to store data
    data = []
    
    # Load and combine data from all JSON files in the specified folder
    for file_path in glob.glob(os.path.join(folder_path, "log_until_*.json")):
        with open(file_path, 'r') as f:
            data.extend(json.load(f))
    
    # Convert to DataFrame
    df = pd.DataFrame(data)
    
    # Sort by step for correct plotting
    df = df.sort_values(by="step")

    # Ensure output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Plot each metric separately and save
    plt.figure(figsize=(10, 5))
    plt.plot(df["step"], df["num_nodes"], label="Number of Nodes")
    plt.xlabel("Step")
    plt.ylabel("Number of Nodes")
    plt.legend()
    plt.title("Number of Nodes Over Steps")
    plt.grid(True)
    plt.savefig(os.path.join(output_folder, "visualize_num_nodes.png"))
    plt.show()
    plt.close()

    plt.figure(figsize=(10, 5))
    plt.plot(df["step"], df["num_edges"], label="Number of Edges", color="orange")
    plt.xlabel("Step")
    plt.ylabel("Number of Edges")
    plt.legend()
    plt.title("Number of Edges Over Steps")
    plt.grid(True)
    plt.savefig(os.path.join(output_folder, "visualize_num_edges.png"))
    plt.show()
    plt.close()

    plt.figure(figsize=(10, 5))
    plt.plot(df["step"], df["mse"], label="MSE", color="green")
    plt.xlabel("Step")
    plt.ylabel("MSE")
    plt.legend()
    plt.title("MSE Over Steps")
    plt.grid(True)
    plt.savefig(os.path.join(output_folder, "visualize_mse.png"))
    plt.show()
    plt.close()

    plt.figure(figsize=(10, 5))
    plt.plot(df["step"], df["success_rate"], label="Success Rate", color="red")
    plt.xlabel("Step")
    plt.ylabel("Success Rate")
    plt.legend()
    plt.title("Success Rate Over Steps")
    plt.grid(True)
    plt.savefig(os.path.join(output_folder, "visualize_success_rate.png"))
    plt.show()
    plt.close()

# Usage example (commented out because the function requires a specific folder and output path):
# load_and_save_logs('/path/to/folder', '/path/to/output_folder')
```
