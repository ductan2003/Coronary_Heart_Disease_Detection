```python
import numpy as np
import matplotlib.pyplot as plt

# Function to create a single ECG wave
def ecg_wave(t, amp=1, width=0.15, spacing=0.6):
    wave = np.zeros_like(t)
    for i in range(len(t)):
        if t[i] % spacing < width:
            wave[i] = amp * np.exp(-((t[i] % spacing - width / 2) ** 2) / (2 * (width / 6) ** 2))
    return wave

# Define time range
t = np.linspace(0, 6, 1000)

# Generate ECG waveform
p_wave = ecg_wave(t, amp=0.3, width=0.2, spacing=1)
q_wave = -0.1 * ecg_wave(t + 0.04, amp=1, width=0.1, spacing=1)
r_wave = ecg_wave(t + 0.08, amp=1, width=0.05, spacing=1)
s_wave = -0.15 * ecg_wave(t + 0.12, amp=1, width=0.1, spacing=1)
t_wave = 0.3 * ecg_wave(t + 0.2, amp=1, width=0.3, spacing=1)

# Combine waves to mimic an ECG
ecg = p_wave + q_wave + r_wave + s_wave + t_wave

# Generate random R-peak offsets
np.random.seed(42)  # Set seed for reproducibility
num_r_peaks = 6
random_offsets = np.random.uniform(-0.05, 0.05, size=num_r_peaks)  # Timing offsets
random_amplitudes = np.random.uniform(0.9, 1.1, size=num_r_peaks)  # Amplitude variability

# Add random variability to R-peak annotations
plt.figure(figsize=(12, 4))
plt.plot(t, ecg, color="black", linewidth=2)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("ECG Waveform with Randomized R-Peaks")
plt.grid(False)

for i, offset in enumerate(random_offsets):
    # Modify R-peak amplitude and timing
    peak_time = i + 0.08 + offset
    peak_amplitude = 1 * random_amplitudes[i]
    if (i == 1):
        peak_time = 1.78
    
    # Annotate the R-peak
    plt.annotate("R", xy=(peak_time, peak_amplitude), xytext=(peak_time, peak_amplitude + 0.2),
                 arrowprops=dict(facecolor='black', arrowstyle="->"),
                 ha='center')

plt.tight_layout()
plt.show()
```
